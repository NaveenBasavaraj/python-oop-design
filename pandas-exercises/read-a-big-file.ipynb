## Reading a big file

```python
import pandas as pd
```

```python
airline_df = pd.read_csv("Airline_Delay_Cause.csv")
```

```python
airline_df.head()
```

```python
cols = [
    "year", "month", "carrier", "airport",
    "arr_flights", "arr_del15", "arr_cancelled",
    "carrier_ct", "weather_ct", "nas_ct", "security_ct", "late_aircraft_ct",
    "carrier_delay", "weather_delay", "nas_delay", "security_delay", "late_aircraft_delay"
]
```

```python
results = []
```

```python
for chunk in pd.read_csv('Airline_Delay_Cause.csv', chunksize=100_000, usecols=cols):
    grouped = chunk.groupby(['carrier']).sum(numeric_only=True)
    results.append(grouped)
results
```

```python
summary = pd.concat(results).groupby('carrier').sum()
print("summary type", type(summary))
print(summary)
```

```python
summary['percent_cancelled'] = summary['arr_cancelled'] / summary['arr_flights'] * 100
summary['percent_delayed'] = summary['arr_del15'] / summary['arr_flights'] * 100
print(summary)
```

```python
summary.to_csv('airline_delay_or_cancelled_causes.csv', index=True, encoding='utf-8')
```
