## Decorator Exercises

1. Create a logger function that logs function name and execution time.

```python
import time

def logging_decorator(func):
    def wrapper(*args, **kwargs):
        print(f"Entering {func.__name__} with args: {args} and kwargs: {kwargs}")
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        execution_time = end_time - start_time
        print(f"{func.__name__} took {execution_time:.4f} and returned: {result}")
        return result
    return wrapper 

@logging_decorator
def calculate_multiply(numbers):
    tot = 1
    for x in numbers:
        tot *= x
    return tot

result = calculate_multiply([1, 2, 3, 4, 5])
print("Result:", result)
```
